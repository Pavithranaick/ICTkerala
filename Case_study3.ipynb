```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Add_sales_data_2018.csv')
data
```

```python
data.head()
```

1. First of all, check whether there is any relationship between the amount spent for advertising through five different companies and profit generated.

```python
correlation_data = data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(correlation_data,fmt='.4g',vmin=-2,vmax=1,annot=True,linewidths=0.2,cmap='Oranges')
```

2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
data.columns
```

```python
y= data['Profit(in Thousands)']
X= data.drop(['Date','Profit(in Thousands)'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=43, test_size = 0.25)

from sklearn import linear_model

lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

prediction_model=model.predict(X_test)
```

```python
print(y_test)
```

```python
print(prediction_model)
```

```python
from sklearn.metrics import mean_squared_error

print('MSE is : ',mean_squared_error(y_test,prediction_model))
```

3. Predict the profit generated by the company, if they wish to spend the following amount for advertisement.

```python
make_prediction = [[290,312,309,298,320]]
model.predict(make_prediction).item()
```

Summary:a strong positive correlation between the amount spent for advertising and profit generated in all the five companies. This means that how much ever money they have spent for advertising that much increase in profit is generated. This prediction indicates the profit generated, if the 5 companies wish to spend the given amount for advertisement then the profit generated is Rs.5820.06.
